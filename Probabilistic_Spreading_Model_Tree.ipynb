Probabilistic Spreading Model of epidemic in a tree  structure 

x is the probability that a node at level h-1 is infected . f(x) is the probability that a node at level h is infected. It is assumed to be a tree structure where node at each level has d children,and its called degree d. q is the infection probability that node of level h-1 can infect a node from level h. Initially at 1st level x is taken to be infected and its probability is 1   

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
def func(x,d):
    #q is kept fixed
    print("x=",x)
    q = 0.5
    return 1-(1- q*x)**d
```

f(x) = 1-(1-q*x)^d 


f'(x) = q*d(1-q*x)^(d-1)

This derivative is a non increasing function

f'(0) = q*d

If we want an epidemic to die out then f(x) must go to 0. Its graph must lie below y=x

```python
x = np.arange(1.0,0.0,-0.1)
```

```python
# y = x
# f = func(x,d)
# print(f)
# plt.figure()
# plt.xticks(x)
# plt.plot(x,y,'r')
# plt.plot(x,f,'b')
# plt.show()
```

```python
for d in range(1,6):
    y = x
    f = func(x,d)
    print("d = %d"%d)
    print("f(x)=",f)
    plt.figure()
    plt.xticks(x)
    plt.plot(x,y,'r')
    plt.plot(x,f,'b')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.show()
```

q value is kept fixed at 0.5 and degree d is varied from 1 to 5.We can see if  degree is 1 and 2 the graph of function is below y=x. In other cases it exceedes y=x.Through social distancing we degree d could be reduced.

```python
def func2(x,q):
    #d is kept fixed
    print("x=",x)
    d = 2
    return 1-(1- q*x)**d
```

```python
for q in np.arange(0.1,1.0,0.1):
    y = x
    f = func2(x,q)
    print("q = %f"%q)
    print("f(x)=",f)
    plt.figure()
    plt.xticks(x)
    plt.plot(x,y,'r')
    plt.plot(x,f,'b')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.show()
```

Here degree d is fixed at 2 ie at each level one node is connected at 2 nodes at its lower level. It is like a binary tree. value of q is varied from 0.1 to 1.0. Here we find if q value is upto 0.5 , the function curve remains below y=x line.At 0.6 it almost gets very close and as q increases , the function overshoots. q needs to be reduced through proper hygine , washing hands , following guidelines.

q*d is average number of nodes, one node can infect. It is also called R0 , Reproductive Number

```python
def func3(x,d,q):
    print("x=",x)
    return 1-(1- q*x)**d
```

```python
for d in range(2,5):
    for q in np.arange(0.4,0.9,0.1):
        y = x
        print("d = %f"%d)
        print("q = %f"%q)
        #print("q*d= %f"%q*d)
        print("R0 = %f"% (q*d))
        f = func3(x,d,q)
        print("f(x)=",f)
        plt.figure()
        plt.xticks(x)
        plt.plot(x,y,'r')
        plt.plot(x,f,'b')
        plt.xlabel('x')
        plt.ylabel('f(x)')
        plt.show()
```
